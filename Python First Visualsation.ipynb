Function

All the above should be generated automatically when we give the name of data table we want to see and the name of players we are interested in regarding the selected variables. First making possible to select names.


```python
def myplot(playerlist):
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c='Black', ls= '--', marker = 's', label = name)
        plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.legend(loc='upper left', bbox_to_anchor = (1,1))
    plt.show()
```

```python
myplot(['KobeBryant', 'LeBronJames', 'DerrickRose'])
```

```python
Colors = {"KobeBryant":'Black',"JoeJohnson":'Red',"LeBronJames":'Green',"CarmeloAnthony":'Blue',"DwightHoward":'Magenta',"ChrisBosh":'Cyan',"ChrisPaul":'Yellow',"KevinDurant":'Black',"DerrickRose":'Red',"DwayneWade":'Blue'}
Linestyles = {"KobeBryant":'--',"JoeJohnson":'--',"LeBronJames":'-.',"CarmeloAnthony":':',"DwightHoward":':',"ChrisBosh": '-,',"ChrisPaul":'-',"KevinDurant":'-',"DerrickRose":'-',"DwayneWade":'-'}
Markers = {"KobeBryant":'s',"JoeJohnson":'o',"LeBronJames":'^',"CarmeloAnthony":'',"DwightHoward":'D',"ChrisBosh":'x',"ChrisPaul":'h',"KevinDurant":'d',"DerrickRose":'H',"DwayneWade":'.'}

def myplot(playerlist):
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c=Colors[name], ls= Linestyles[name], marker = Markers[name], label = name)
        plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.legend(loc='upper left', bbox_to_anchor = (1,1))
    plt.show()
```

Fix up the input: we should be able to decide which data sets we want to see. Thus myplot() function sould have two variables:
data and playerlist.
We set the whole Players dictionary as default to playerlist variable. In case we don't choose a name, vis will show all players' dataset.

```python
Colors = {"KobeBryant":'Black',"JoeJohnson":'Red',"LeBronJames":'Green',"CarmeloAnthony":'Blue',"DwightHoward":'Magenta',"ChrisBosh":'Cyan',"ChrisPaul":'Yellow',"KevinDurant":'Black',"DerrickRose":'Red',"DwayneWade":'Blue'}
Linestyles = {"KobeBryant":'--',"JoeJohnson":'--',"LeBronJames":'-.',"CarmeloAnthony":':',"DwightHoward":':',"ChrisBosh": '-.',"ChrisPaul":'-',"KevinDurant":'-',"DerrickRose":'-',"DwayneWade":'-'}
Markers = {"KobeBryant":'s',"JoeJohnson":'o',"LeBronJames":'^',"CarmeloAnthony":'x',"DwightHoward":'D',"ChrisBosh":'x',"ChrisPaul":'h',"KevinDurant":'d',"DerrickRose":'H',"DwayneWade":'.'}

def myplot(data, playerlist=Players):
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Colors[name], ls= Linestyles[name], marker = Markers[name], label = name)
        plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.legend(loc='upper left', bbox_to_anchor = (1,1))
    plt.show()
```
